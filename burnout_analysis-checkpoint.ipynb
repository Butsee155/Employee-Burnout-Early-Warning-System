```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/employee_burnout_synthetic_dataset.csv")
df.head()
```

```python
df["calculated_burnout_risk"] = (
    (df["hours_worked"] - 8) * 6 +
    df["meetings_count"] * 2.5 +
    df["task_switches"] * 1.3 -
    df["break_minutes"] * 0.25 +
    (60 - df["output_score"])
)

df["calculated_burnout_risk"] = df["calculated_burnout_risk"].clip(0, 100)
```

```python
def risk_level(score):
    if score < 30:
        return "Low"
    elif score < 60:
        return "Medium"
    else:
        return "High"

df["burnout_level"] = df["calculated_burnout_risk"].apply(risk_level)
```

```python
sns.histplot(df["calculated_burnout_risk"], bins=30, kde=True)
plt.title("Distribution of Burnout Risk Scores")
plt.show()
```

```python
daily_burnout = df.groupby("day")["calculated_burnout_risk"].mean()

plt.figure(figsize=(10,5))
plt.plot(daily_burnout)
plt.title("Average Burnout Risk Over Time")
plt.xlabel("Day")
plt.ylabel("Burnout Risk")
plt.show()
```

```python
corr = df[[
    "hours_worked",
    "meetings_count",
    "task_switches",
    "break_minutes",
    "output_score",
    "calculated_burnout_risk"
]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation with Burnout Risk")
plt.show()
```

```python
sample_emp = df[df["employee_id"] == 25]

plt.plot(sample_emp["day"], sample_emp["calculated_burnout_risk"])
plt.title("Burnout Trend for Employee 25")
plt.xlabel("Day")
plt.ylabel("Burnout Risk")
plt.show()
```
